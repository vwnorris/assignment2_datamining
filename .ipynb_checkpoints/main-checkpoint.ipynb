```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import RidgeCV, Ridge
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import classification_report
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score, classification_report, accuracy_score
```

```python
# Load the dataset
df = pd.read_csv('water_potability2.csv')

# Remove rows with NaN values
df_cleaned = df.dropna()
print(df_cleaned.head())

# Define the feature matrix X and target vector y after cleaning
X_cleaned = df_cleaned.drop('Potability', axis=1)
y_cleaned = df_cleaned['Potability']
```

```python
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit the scaler to the cleaned data and transform it
X_cleaned_scaled = scaler.fit_transform(X_cleaned)

# Split the cleaned and scaled data into training and testing sets
X_train_cleaned, X_test_cleaned, y_train_cleaned, y_test_cleaned = train_test_split(
    X_cleaned_scaled, y_cleaned, test_size=0.2, random_state=42, shuffle=True
)
```

```python
# Create and train the Linear Regression model using the cleaned and scaled data
linear_model = LinearRegression()
linear_model.fit(X_train_cleaned, y_train_cleaned)

# Make predictions using the testing set
y_pred_cleaned = linear_model.predict(X_test_cleaned)
```

```python
# Print the coefficients and regression metrics
print('Coefficients:', linear_model.coef_)
print('Mean squared error: %.2f' % mean_squared_error(y_test_cleaned, y_pred_cleaned))
print('Coefficient of determination: %.2f' % r2_score(y_test_cleaned, y_pred_cleaned))

# Since the output is continuous and we are dealing with a classification problem,
# We should convert the predictions to a binary outcome
y_pred_binary = (y_pred_cleaned > 0.5).astype(int)

# Print classification metrics with zero_division set to 0
print(classification_report(y_test_cleaned, y_pred_binary, zero_division=0))
print('Accuracy: %.2f' % accuracy_score(y_test_cleaned, y_pred_binary))
```

```python
# Define a range of lambda (alpha) values to test
lambdas = np.logspace(-4, 4, 500)
mse_scores = []

# Compute 10-fold cross-validation for each value of lambda (alpha)
for alpha in lambdas:
    model = Ridge(alpha=alpha)
    scores = cross_val_score(model, X_cleaned_scaled, y_cleaned, scoring='neg_mean_squared_error', cv=10)
    mse_scores.append(-scores.mean())  # Append the mean score
```

```python
# Plotting the MSE scores as a function of lambda
plt.figure()
plt.semilogx(lambdas, mse_scores, label='Cross-validated MSE')
plt.xlabel('Lambda')
plt.ylabel('Mean Squared Error')
plt.title('MSE as a function of Lambda')
plt.legend()
plt.show()

# Find the lambda that minimizes the MSE
min_mse = min(mse_scores)
optimal_lambda = lambdas[mse_scores.index(min_mse)]

# Print the results
print("The optimal lambda found is:", optimal_lambda)
print("The minimum cross-validated MSE is:", min_mse)
```

```python
# Since LogisticRegressionCV applies cross-validation directly, we do not need the loop
# Cs is the inverse of regularization strength; smaller values specify stronger regularization
Cs = np.logspace(-4, 4, 100)

# Create logistic regression model with cross-validation
logistic_model_cv = LogisticRegressionCV(
    Cs=Cs, cv=10, scoring='accuracy', penalty='l2', solver='liblinear', max_iter=1000
)

# Fit model
logistic_model_cv.fit(X_cleaned_scaled, y_cleaned)

# Print the optimal C (which is the inverse of lambda)
optimal_C = logistic_model_cv.C_[0]
print("The optimal C found is:", optimal_C)

# You can now evaluate the model using the score or any other evaluation metric
print(classification_report(y_cleaned, logistic_model_cv.predict(X_cleaned_scaled)))
```
